```python
import numpy as np
# 创建一个符合正态分布的10个股票5天的涨跌幅数据
stock_change = np.random.normal(0, 1, (10, 5))
```

```python
stock_change # display data 
# 我们不知道这些数据代表什么意思, 所以我们需要给他加上行列索引
# 也就是我们的DataFrame数据类型
```

```python
import pandas as pd # import package
```

```python
pd.DataFrame(stock_change) 
# 我们没有指定索引, 所以它默认就是0~...
```

```python
# 添加行索引
stock = ["股票{}".format(i) for i in range(10)]
```

```python
pd.DataFrame(stock_change, index=stock)
# index: 行索引
```

```python
# 添加列索引
# date_range是专门处理日期的函数
date = pd.date_range(start="20180101", periods=5, freq="B")
```

```python
date
```

```python
data = pd.DataFrame(stock_change, index=stock, columns=date)
# 成功将列索引添加上去
```

```python
data
```

### 设置新索引

```python
# 我们还可以使用字典来生成我们的DataFrame
df = pd.DataFrame({'month': [1, 4, 7, 10],
                    'year': [2012, 2014, 2013, 2014],
                    'sale':[55, 40, 84, 31]})
```

```python
df
```

```python
# 以月份设置新的索引
# 只有一个索引, 设置为新的索引, 并且drop=True删除原来的列.
df.set_index("month", drop=True)
```

```python
# 设置多个索引，以年和月份
# 将多个索引值放入一个列表, 传入进去
new_df = df.set_index(["year", "month"])
```

```python
new_df.index
# 当我们设置多个索引的时候,
# 返回的就是MultiIndex了, 此时数据就是三维数据了
```

```python
new_df.index.names # 索引的名字
```

```python
new_df.index.levels # 名字里面的
```

### 1 逻辑运算符号<、 >、|、 &

```python
# 例如筛选p_change > 2的日期数据
data.pchange>2 # -> 返回的是带索引的一组bool值
data[data["p_change"] > 2].head() # 然后做bool索引
```

```python
# 完成一个多个逻辑判断， 筛选p_change > 2并且low > 15
# 获取逻辑结果, 然后做bool索引
data[(data["p_change"] > 2) & (data["low"] > 15)].head()
```

```python
# query(expr)
# expr: 直接写查询的字符串即可, 非常简单
data.query("p_change > 2 & low > 15").head() # 
```

```python
# 判断'turnover'是否有4.19, 2.39
# 返回也是一个逻辑矩阵, 然后我们也可以做布尔索引
data[data["turnover"].isin([4.19, 2.39])]  
```
